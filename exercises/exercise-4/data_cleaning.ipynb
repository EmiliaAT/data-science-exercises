```python
import pandas as pd
import numpy as np

# Set random seed for reproducibility
np.random.seed(42)

# Generate 50 normal data points
data = pd.DataFrame({'Value': np.random.normal(loc=50, scale=5, size=50)})

# Insert some outliers
outliers = pd.DataFrame({'Value': [100, 110, 120, 130]})

# Append data with outliers
data_with_outliers = pd.concat([data, outliers], axis=0)

# Create DataFrame
df = pd.DataFrame(data_with_outliers, columns=['Value'])

print(df.describe())
print()

df['Z-Score'] = (df['Value'] - df['Value'].mean()) / df['Value'].std()

print(df['Z-Score'])
print()

found_outliers = df[df['Z-Score'].abs() > 3]

print(found_outliers)
print()

with_log = np.log1p(df['Value'])

print(with_log)
print()

clean_df = df[df['Z-Score'].abs() <= 3][['Value']]

print(clean_df.describe())
print()

print(clean_df)
print()
```
